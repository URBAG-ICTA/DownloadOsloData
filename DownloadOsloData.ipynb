```python
import pandas as pd
import io
import requests
import json
```

```python
base_url = "https://api.nilu.no"
```

```python
air_quality_index_url = "/aq/historical/{fromtime}/{totime}/{latitude}/{longitude}/{radius}?components={components}"
```

```python
https://api.nilu.no/aq/historical/2017-01-01/2017-01-02/59.89869/10.81495/3.0?components=no2;pm10
```

```python
start_date = "2017-01-01"
end_date = "2017-01-02"
lat = "59.89869"
lon = "10.81495"
rad = "3.0"
comp = "no2;pm10"
```

```python
complete_url = base_url+air_quality_index_url.format(fromtime = start_date, totime = end_date, latitude = lat, longitude = lon, radius=rad, components = comp)
complete_url
```

```python
#from url to pandas dataframe
complete_url = base_url+air_quality_index_url.format(fromtime = start_date, totime = end_date, latitude = lat, longitude = lon, radius=rad, components = comp)
print(complete_url)

raw_data=requests.get(complete_url).content
raw_data_decoded = raw_data.decode('utf-8')
res = json.loads(raw_data_decoded)
res
df = pd.DataFrame.from_dict(res[0])
```

```python
df['fromTime'] = df.apply(lambda r: r['values']['fromTime'], axis=1)
df['toTime'] = df.apply(lambda r: r['values']['toTime'], axis=1)
df['value'] = df.apply(lambda r: r['values']['value'], axis=1)
df['qualityControlled'] = df.apply(lambda r: r['values']['qualityControlled'], axis=1)
df['index'] = df.apply(lambda r: r['values']['index'], axis=1)
df['color'] = df.apply(lambda r: r['values']['color'], axis=1)
```
